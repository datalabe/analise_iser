```python
def count(df, column):
    df_novo = pd.DataFrame(df[column].value_counts())
    df_novo_percent = pd.DataFrame(df_novo * 100 / df_novo.sum())
    df_comp = df_novo_percent.join(df_novo, lsuffix='%')
    df_comp.loc['Total'] = df_comp.sum()
    
    return  round(df_comp)
```
